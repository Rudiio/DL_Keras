```python
import keras
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

# Final Assignment

##   Â Loading the data



```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
predictors = concrete_data.drop("Strength",1)
target = concrete_data["Strength"]

input_shape = predictors.shape[1]
```

## A. Build a baseline model

### Building the model


```python
def RegressionModel():
    model=Sequential()
    model.add(Dense(10,activation="relu",input_shape=(input_shape,)))
    model.add(Dense(1))
    
    model.compile(optimizer="adam",loss="mean_squared_error", metrics=["mean_squared_error"])

    return model
```


###   Training process


```python
scores = []

for i in range(0,50):
    print(f"----------> Step {i}/50")

    model = RegressionModel()
    x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size=0.3)
    model.fit(x_train,y_train,epochs=50,verbose=2)
    
    predictions = model.predict(x_test)
    mse = mean_squared_error(y_test,predictions)

    scores.append(mse)
    print(mse)
```

```python
print(f"Mean of mean squared errors = {np.mean(scores)}")
print(f"Standard deviation of mean squared errors = {np.std(scores)}")
```

The results are not very interesting as we got a big means of MSE that is way too large in comparison to our the range of the data.
The same statement applies to the variation of the results according to the standard deviation.

## B. Normalize the data

We normalize the data.

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

### Training process

```python
scores_norm = []

for i in range(0,50):
    print(f"----------> Step {i}/50")

    model = RegressionModel()
    x_train,x_test,y_train,y_test = train_test_split(predictors_norm,target,test_size=0.3)
    model.fit(x_train,y_train,epochs=50,verbose=2)
    
    predictions = model.predict(x_test)
    mse = mean_squared_error(y_test,predictions)

    scores_norm.append(mse)
```

```python
print(f"Mean of mean squared errors with normalize data = {np.mean(scores_norm)}")
print(f"Standard deviation of mean squared errors with normalize data = {np.std(scores_norm)}")
```

The results obtained with normalized is less interesting according to the mean of MSEs but it has less variations has indicated by the lower standard deviation.

##     C. Increase the number of epochs

```python
scores_norm_100 = []

for i in range(0,50):
    print(f"----------> Step {i}/50")

    model = RegressionModel()
    x_train,x_test,y_train,y_test = train_test_split(predictors_norm,target,test_size=0.3)
    model.fit(x_train,y_train,epochs=100,verbose=2)
    
    predictions = model.predict(x_test)
    mse = mean_squared_error(y_test,predictions)

    scores_norm_100.append(mse)
```

```python
print(f"Mean of mean squared errors with normalize data and 100 epochs = {np.mean(scores_norm_100)}")
print(f"Standard deviation of mean squared errors with normalize data and 100 epochs = {np.std(scores_norm_100)}")
```

We get way better result when increasing the number of epochs. The mean of MSE is almost halfed and the standard deviation is way smaller so we get less variations.

## D. Increase the number of hidden layers 

```python
def LargerRegressionModel():
    model=Sequential()
    model.add(Dense(10,activation="relu",input_shape=(input_shape,)))
    model.add(Dense(10,activation="relu"))
    model.add(Dense(10,activation="relu"))
    model.add(Dense(1))
    
    model.compile(optimizer="adam",loss="mean_squared_error", metrics=["mean_squared_error"])

    return model
```

```python
scores_large = []

for i in range(0,50):
    print(f"----------> Step {i}/50")

    model = LargerRegressionModel()
    x_train,x_test,y_train,y_test = train_test_split(predictors_norm,target,test_size=0.3)
    model.fit(x_train,y_train,epochs=50,verbose=2)
    
    predictions = model.predict(x_test)
    mse = mean_squared_error(y_test,predictions)

    scores_large.append(mse)
```

```python
print(f"Mean of mean squared errors with larger model = {np.mean(scores_large)}")
print(f"Standard deviation of mean squared errors with larger model = {np.std(scores_large)}")
```

Modifying the model architecture by adding more hidden layers also increase the performance in comparison to part B as showed by the resulting mean of MSE that is way smaller. So it is necessary to pay attention to the model architecture.
